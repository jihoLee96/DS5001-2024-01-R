   The method used for calculating TFIDF is 'max' for TF, and 'standard' for IDF.
   The calculation normalizes TF by dividing each term's frequency by the maximum frequency of any term in the document.
   IDF is calculated by taking the logarithm of the total number of documents divided by the DF of each term.
    
   DTCM = BOW.n.unstack() \
   TF = (DTCM.T / DTCM.T.max()).T #max \
   IDF = np.log2(N_docs/DF) #standard \
   TFIDF = TF*IDF \
   TFIDF = TFIDF.fillna(0)
