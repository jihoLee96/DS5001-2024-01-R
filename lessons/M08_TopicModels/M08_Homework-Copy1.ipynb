````python
# Set Up and Imports
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation as LDA
import plotly_express as px

# Config
import configparser
config = configparser.ConfigParser()
config.read('../../../env.ini')
data_home = config['DEFAULT']['data_home']
output_dir = config['DEFAULT']['output_dir']

data_prefix = 'novels'
colors = "YlGnBu"

ngram_range = (1, 2) # (1,2) 였음
n_terms = 4000
n_topics = 20
max_iter = 5
n_top_terms = 7

OHCO = ['book_id', 'chap_id', 'para_num']
PARA = OHCO[:3]
CHAP = OHCO[:2]
BOOK = OHCO[:1]

BAG = CHAP

## Pragmas
import warnings
warnings.filterwarnings('ignore')
````

````python
# Preapare data

# Conver TOKENS back to DOCS

TOKENS = pd.read_csv(f'{data_home}/{data_prefix}-CORPUS.csv')

TOKENS
# BAG
````

````python
def gen_model(TOKENS, BAG):
    # Filter for Nouns
    DOCS = TOKENS[TOKENS.pos.str.match(r'^NNS?$')]\
        .groupby(BAG).term_str\
        .apply(lambda x: ' '.join(x))\
        .to_frame()\
        .rename(columns={'term_str':'doc_str'})
    # print(DOCS.head())

    # Create vector space
    # F1 corpus of paragraphs into a document-term vector sapce of word counts
    count_engine = CountVectorizer(max_features=n_terms, ngram_range=ngram_range, stop_words='english')
    count_model = count_engine.fit_transform(DOCS.doc_str)
    TERMS = count_engine.get_feature_names_out()

    VOCAB = pd.DataFrame(index=TERMS)
    VOCAB.index.name = 'term_str'

    DTM = pd.DataFrame(count_model.toarray(), index=DOCS.index, columns=TERMS)
    # print(DTM)

    VOCAB['doc_count'] = DTM.astype('bool').astype('int').sum()
    DOCS['term_count'] = DTM.sum(1)
    # print(VOCAB)
    # print(DOCS.term_count.describe())
    
   # Generate Model
    # We run Scikit Learn's LatentDirichletAllocation algorithm and extract the
    # THETA and PHI tables.

    lda_engine = LDA(n_components=n_topics, max_iter=max_iter, learning_offset=50., random_state=0)

    # Topic names
    TNAMES = [f"T{str(x).zfill(len(str(n_topics)))}" for x in range(n_topics)]
 
    # THETA
    lda_model = lda_engine.fit_transform(count_model)

    THETA = pd.DataFrame(lda_model, index=DOCS.index)
    THETA.columns.name = 'topic_id'
    THETA.columns = TNAMES
    
    # PHI
    PHI = pd.DataFrame(lda_engine.components_, columns=TERMS, index=TNAMES)
    PHI.index.name = 'topic_id'
    PHI.columns.name = 'term_str'
#     PHI.T.sample(10).style.background_gradient(cmap=colors, axis=None)

    # Inspect Results
    # Create TOPICS and get Top terms per Topic
    TOPICS = PHI.stack().groupby('topic_id')\
        .apply(lambda x: ' '.join(x.sort_values(ascending=False).head(n_top_terms).reset_index().term_str))\
        .to_frame('top_terms')
    
    return THETA, PHI, DOCS, TOPICS, lda_engine, lda_model, count_engine, count_model, TNAMES
````

````python
theta_chap, phi_chap, docs_chap, topics_chap, lda_engine_chap, lda_model_chap, count_engine_chap, count_model_chap, tnames_chap = gen_model(TOKENS, BAG)
````

````python
theta_para, phi_para, docs_para, topics_para, lda_engine_para, lda_model_para, count_engine_para, count_model_para, tnames_para = gen_model(TOKENS, PARA)
````

````python
# # to compare the entropies of the topic distributions between paragraphs and chapters
# # to identify which generates a lower entropy distribution. 
# # This function provides the average entropy across all topics, 
# # giving a measure of the overall distribution's spread or concentration.

# def compute_H(PHI):
# #     Sum the word weights for each topic
#     topic_sums = PHI.abs().sum(axis=1)
# #     Normalize to get a probability distribution over topics
#     topic_distribution = topic_sums / topic_sums.sum()
# #     Compute the entropy
#     entropay = -np.sum(topic_distribution * np.log2(topic_distribution))
#     return topic_sums, topic_distribution, entropy
# a, b, entropy_chap = compute_H(phi_chap)
# a, b, entropy_para = compute_H(phi_para)
# print(entropy_chap)
# print(entropy_para)
````

````python
def compute_H(PHI):
    df_normalized = PHI.div(PHI.abs().sum(axis=1), axis=0)
    # Then, calculate the entropy for each topic
    entropy = (-df_normalized * np.log2(df_normalized)).sum(axis=1, skipna=True)
    return df_normalized, entropy
````

````python
def topic_sort(TOPICS, THETA, PHI):
    # Sort topics by Doc weight
    TOPICS['doc_weight_sum'] = THETA.sum()
    TOPICS['term_freq'] = PHI.sum(1) / PHI.sum(1).sum()
    topic_sort = TOPICS.sort_values('doc_weight_sum', ascending=False).style.background_gradient(cmap=colors)
    return topic_sort
````
